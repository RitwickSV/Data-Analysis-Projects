# Project Goals

We have come across the data for various automobiles and will be performing a data analysis and correlation for various columns.<br>
We first start by importing all the modules required for the analysis.

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

<br>
Next we load the data into a dataframe and see if it is valid enough to be analysed.
<br>

```python
df = pd.read_csv('auto.csv')
display(df.head())
```

<br>We observe that there seems to be no headers for the given data and we also have different values in the dataframe which have `?` as value which needs to be addressed.<br> We start by naming the headers/columns for the data.

```python
df.columns = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style",\
              "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type",\
              "num-of-cylinders", "engine-size","fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm",\
              "city-mpg", "highway-mpg", "price"]
df = df.replace('?',np.NaN)
display(df.head())
```

<br>
Now that we have replaced all the `?` in the data with a NaN value, now it is time to deal with the missing data in the datafraem. But before that, we need to find all the columns which have the missing values.

```python
null_col_list = []
for column in df.columns.values:
    if(any(df[column].isna())):
        null_col_list.append(column)
    else:
        continue
print(null_col_list)
```

<br> Now that we have found the columns having the missing value, we can proceed how to deal with the data. We observe that we can replace all the missing values with their resepctive column's mean value except for `num-of-doors`. So we replace the missing doors with 4 and the rest of the values with the mean.

```python
df['normalized-losses'].replace(np.NaN, df['normalized-losses'].astype('float').mean(axis = 0), inplace = True)
df['bore'].replace(np.NaN, df['bore'].astype('float').mean(axis = 0), inplace = True)
df['horsepower'].replace(np.NaN, df['horsepower'].astype('float').mean(axis = 0), inplace = True)
df['peak-rpm'].replace(np.NaN, df['peak-rpm'].astype('float').mean(axis = 0), inplace = True)
df['price'].replace(np.NaN, df['price'].astype('float').mean(axis = 0), inplace = True)
df['stroke'].replace(np.NaN, df['stroke'].astype('float').mean(axis = 0), inplace = True)
```

<br>
Now that we have taken care of the missing values in the dataframe, let's observe the datatypes of the columns and convert them to the appropriate data types.

```python
df.dtypes
```

```python
df[['bore', 'stroke', 'price', 'peak-rpm']] = df[['bore', 'stroke', 'price', 'peak-rpm']].astype('float')
df[['normalized-losses', 'horsepower']] = df[['normalized-losses', 'horsepower']].astype('int')
```

```python
df.dtypes
```

```python
display(df.head())
```

<br>
Let us compare how the price values fare as compared with the engine-size, highway mileage, peak-rpm and various other factors of the car

```python
def fx_col_price_plot(col, data):
    sns.regplot(data = data, x = col, y = 'price')
    sns.set_style('darkgrid')
    plt.rc('axes', titlesize=18)
    plt.rc('axes', labelsize=14)
    plt.rc('xtick', labelsize=13)
    plt.rc('ytick', labelsize=13)
    plt.rc('legend', fontsize=13)
    plt.rc('font', size=13)

for item in ['engine-size', 'highway-mpg', 'peak-rpm', 'horsepower']:
    fx_col_price_plot(item, df)
    plt.xlabel('{}'.format(item))
    plt.ylabel('Price of the Car')
    plt.title('{} vs Price of the Car'.format(item))
    plt.show()
```

<br>
Engine size seems to have a positive correlation with the price. The highway mileage seems to have a negative correlation with the price i.e., the price seems to be low when the mileage is too high. There seems to be no perfect distinguishable correlation between the peak-rpm and the price of the car hence no observation can be made regarding the correlation or the causality between the pair.

<br>It would be helpful if we can compare all the columns at once but it is a hard task. Hence we use the `df.corr()` value to get the pearson coefficients of all the columns as shown below:

```python
df.corr()
```

<br>This correlation matrix gives us the correlation value between the paris of all the quantitative variables in the dataframe. By observation, we surmise that the values are redundant and it would help us if we could get rid of either the first half or the second half and would be easier to read if we could add a heatmap to the matrix.

```python
df.horsepower.unique()
```

```python
corr_matrix = df.corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
cmap = sns.diverging_palette(255, 16, s=100, l=50, n=16, center="light", as_cmap=True)
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, mask = mask, center=0, annot=True, fmt='.2f', square=True, cmap=cmap)
plt.show()
```

<br>We can observe from the heatmap that whatever correlation we have obtained earlier for horsepower, highway mileage, rpm and engine size are the same we obtain in the heatmap.
 - `> 0.7` is a positive correlation
 - `< -0.7` is a negative correlation
 - `>-0.7 & <0.7` is a correlation which is almost unrelated

We will now categorise the prices into low range, mid range and high range by splitting the data into 3 equal sets.

```python
low = np.quantile(df.price,0.33)
med = np.quantile(df.price,0.66)
print(low, med)
```

```python
def fx_prc_category(val):
    if val < low:
        return 'low'
    elif (val >= low) & (val < med):
        return 'med'
    elif val > med:
        return 'high'
df['price_category'] = df['price'].apply(lambda x : fx_prc_category(x))
```

```python
low_df = df[df.price_category == 'low'].sort_values(by = 'price').reset_index(drop = True)
med_df = df[df.price_category == 'med'].sort_values(by = 'price').reset_index(drop = True)
high_df = df[df.price_category == 'high'].sort_values(by = 'price').reset_index(drop = True)
```


###  What are the costliest/cheapest [type of car] in the lot?

```python
rate = input("Which car do you prefer? The costliest or the cheapest?").lower()
style = input("\nWhich type of a car do you prefer? [Convertible, Sedan, Hatchback, Hardtop, Wagon, entire]").lower()
price_range = input("\nWhich price range are you looking for? [low, med, high, entire]").lower()
make = input("\nWhich brand are you looking for? ['alfa-romero', 'audi', 'bmw', 'chevrolet', 'dodge', 'honda', \
'isuzu', 'jaguar', 'mazda', 'mercedes-benz', 'mercury',\
'mitsubishi', 'nissan', 'peugot', 'plymouth', 'porsche', 'renault', \
'saab', 'subaru', 'toyota', 'volkswagen', 'volvo', entire"   ).lower()

def fx_price_cars(rate, style, price_range, make):
    if (price_range == 'entire') & (make == 'entire') & (style == 'entire'):
        return df.sort_values(by = 'price').reset_index(drop = True)

    elif(price_range == 'entire') & (make == 'entire') & (style != 'entire'):
        return df[df['body-style'] == style].\
sort_values(by = 'price').reset_index(drop = True)

    elif(price_range == 'entire') & (make != 'entire') & (style == 'entire'):
        return df[df['make'] == make].\
sort_values(by = 'price').reset_index(drop = True)

    elif(price_range == 'entire') & (make != 'entire') & (style != 'entire'):
        return df[(df['body-style'] == style) & (df['make'] == make)].\
sort_values(by = 'price').reset_index(drop = True)

    elif(price_range != 'entire') & (make == 'entire') & (style == 'entire'):
        return df[df['price_category'] == price_range].\
sort_values(by = 'price').reset_index(drop = True)

    elif(price_range != 'entire') & (make == 'entire') & (style != 'entire'):
        return df[(df['price_category'] == price_range) & (df['body-style'] == style)].\
sort_values(by = 'price').reset_index(drop = True)

    elif(price_range != 'entire') & (make != 'entire') & (style == 'entire'):
        return df[(df['price_category'] == price_range) & (df['make'] == make)].\
sort_values(by = 'price').reset_index(drop = True)

    elif (price_range != 'entire') & (make != 'entire') & (style != 'entire'):
        return df[(df['price_category'] == price_range) & (df['body-style'] == style) & (df['make'] == make)].\
sort_values(by = 'price').reset_index(drop = True)
            
result_df = fx_price_cars(rate,style,price_range,make)
if(result_df.empty):
    print('\n\nThere are no cars with the specifications you have mentioned. Please try again.')
else:
    if rate == 'costliest':
        print('\n\nThe {} {}[s] in the {} price range from the {} brand are\n'.format(rate, style, price_range, make))
        display(result_df.tail(5))
    else:
        print('\n\nThe {} {}[s] in the {} price range from the {} brand are\n'.format(rate, style, price_range, make))
        display(result_df.tail(5))
```
