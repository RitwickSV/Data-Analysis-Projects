# Project Goals

To observe the biodiversity in various national parks and analyse the data available to us. We try to answer questions through our data and graphs about the various factors that result in biodiversity in various national parks.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
obs_df = pd.read_csv('observations.csv')
spec_df = pd.read_csv('species_info.csv')
```

```python

display(spec_df)
display(obs_df)
```

`species_info.csv` - contains data about different species and their conservation status
 - **category:** Contains the category of the animal observed.
 - **scientific_name:** The scientific name of the animal.
 - **common_names:** The common name of the animal.
 - **conservation_status:** The current conservation status of the animal <br>
 
`observations.csv` - holds recorded sightings of different species at several national parks for the past 7 days.
 - **scientific_name:** The scientific name of the species observed at the national park.
 - **park_name:** The name of the park where the observation was made.
 - **observations:** The number of observations of the particular species in the national park.

By using `spec_df.scientific_name.value_counts()` we can further observe that there seems to be duplicates in the species dataframe which seems to be present due to wrong input of the common names where the common names are repeated.

We can get the list of all the duplicate scientific names by using the following code <br>`list(spec_df.scientific_name.value_counts().loc[lambda x : x > 1].reset_index()['index'])` <br> or we could remove the duplicates appropriately by joining the redundant common name values.

We have deployed the `.join()` function to the common names which are redundant.


```python
#Joining the redundant common name values into one row value which can later be obtained by using set() and split() funcs
spec_df['common_names'] = spec_df.groupby(['category', 'scientific_name'])['common_names'].transform(lambda x : ' '.join(x))

#removing the duplicates which we obtain by joining the common names into one single row
spec_df = spec_df.drop_duplicates()
```

<br>
Observe that we haven't used the `conservation_status` column in the group by function because there seems to be NaN values in the columns which needs to be addressed separately.
<br>
We once again check for the duplicates with respect to the conservation_status in the species dataframe
<br>

Upon checking with `spec_df.scientific_name.value_counts()` we observe that the scientific names are again duplicated because there seems to be a NaN value in the reservation_status which we remove by using the<br>
`fillna(method = 'bfill')` after group by

```python
spec_df['conservation_status'] = spec_df.groupby(['category', 'scientific_name'])['conservation_status'].\
apply(lambda x : x.fillna(method = 'bfill'))
spec_df = spec_df.drop_duplicates()
spec_df.reset_index(drop = True, inplace = True)
```

We still seem to have duplicates in the reservation_status for one scientific name **Canis lupus** because there seems to be two wrong entries that the species is both *endangered* and *in recovery*

<br>
We seem to have duplicates and NaN values in the observation dataframes as well.
<br>

```python
#To remove NaN values with 0 in the observations column and removing the duplicates generated
obs_df.observations.fillna(0)
obs_df = obs_df.drop_duplicates()
```

```python
#To aggregate the duplicates row wise upon observations and adding each of them to a new row
obs_df['observations']= obs_df.groupby(['scientific_name', 'park_name'])['observations'].transform(lambda x : sum(x))
```

```python
obs_df = obs_df.drop_duplicates()
obs_df.reset_index(drop = True, inplace = True)
display(obs_df)
```

<br>

## Analysing the columns and observing the correlation<br>
We have tidied the data which is now enough to process our analysis using graphs relating each of the columns.
<br>
We begin by trying to answer the question: <br>
`Which category of the animal seems to be the most endangered or the least?`

<br>

```python
spec_df.fillna('No Risk', inplace = True)
spec_df.reset_index(drop = True, inplace = True)
```

```python
spec_df['conservation_status'] = pd.Categorical(spec_df.conservation_status, \
                                                      ['No Risk', 'In Recovery', 'Species of Concern', 'Threatened', 'Endangered'],\
                                                      ordered = True)
```

```python
spec_df['conservation_value'] = spec_df.conservation_status.cat.codes + 1
```

```python
spec_df = spec_df.drop_duplicates()
```

```python
display(spec_df)
```

```python
plt.figure(figsize = (8,8))
fig = sns.violinplot(x = spec_df.category, y = spec_df.conservation_value)
plt.xlabel('Category of the animal')
plt.xticks(rotation = 90)
plt.ylabel('Conservation status')
#fig.set_yticklabels(['Endangered', 'In Recovery', 'Species of Concern', 'Threatened', 'Endangered'])
plt.show()
```

We observe that the **Fish** and the **Amphibian** category of the species seems to be closer to being endangered than the others. <br>
We could delve deeper into the data as well.

```python
merged_data = pd.merge(obs_df, spec_df, on = 'scientific_name', sort = False)
```

```python
merged_data
```

## How many of the species from each of the category are endangered?

Let's analyse the data with respect to category, number of scientific_names present in the data and their respective conservations status

```python
cat_cons_df = merged_data.groupby(['category','conservation_status']).scientific_name.count().reset_index()
cat_cons_df.rename(columns = {'scientific_name' : 'species_count'}, inplace = True)
display(cat_cons_df)
```

We now plot the barplots of our observations to get an idea about how many of the species are endangered. We use the log scale for depicting the count as it is too high for a few categories

```python
plt.figure(figsize = (14,8))
fig = sns.barplot(data = cat_cons_df, x = 'category', y = 'species_count', hue = 'conservation_status')
fig.set_yscale('log')
plt.xlabel('Category of the species')
plt.ylabel('Number of species')
plt.legend(bbox_to_anchor = (1.1, 0.5))
plt.show()
```

## How is the number of observations related to the parks in which the observations have been made?

```python
plt.figure(figsize = (14,8))
sns.barplot(data = merged_data, x = 'park_name', y = 'observations', hue = 'category')
plt.show()
```
